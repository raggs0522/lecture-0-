```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
x=np.linspace(0,5,11)##linspace creates a list of the arranged numbers of equal space between the 2 given numbers 
y=x**2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL
plt.plot(x,y)
#plt.show()
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("n vs n**2")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented method for creating figure objects 
fig=plt.figure()
#(left,bottom,width,height)

axes=fig.add_axes([0.1,0.1,0.8,0.4])
axes.plot(x,y)
axes.set_xlabel("XLabel")
axes.set_ylabel("YLabel")
axes.set_title("x vs square")
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.2,0.6,0.6])
axes1.plot(x,y)
axes1.set_title("LArger plot")
axes2.plot(y,x,'r')
##all these numbers have to be between 0 and 1 only always as the second ones numbers are relative to that of the first one, for 
##instance, 0.2 means that 20% left of the previous and so on 
```

```python
fig,axes=plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
```

```python
##coming back to the 1,2 subplot
fig,axes=plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
```

```python
## when we create an axes using this thing then we are basically creating a list, and we can easily iterate through this list.
## we can acess all the detils from this usig normal indexing of a list.
fig,axes=plt.subplots(nrows=1,ncols=2)#we donot have to manually create the axes here , what happens here is tupple unpacking basically 
axes[0].plot(x,y)
axes[0].set_xlabel("XLABEL")
axes[0].set_ylabel("YLABEL")
axes[1].plot(y,x)
```

# Matplot lib also allows us to control the figure size and the DPI(Dots per inch)

```python
fig=plt.figure(figsize=(4,4))
ax=fig.add_axes([0,0,1,1])
```

```python
##now the above thing can be done in the following way as well :
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
##Legends
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,label='X vs Y')
ax.plot(y,x,label='Y vs X')
ax.legend(loc=0)#this method helps us to identify the different curves that are present there., the loc parameter will help us situate the legend t the best possible location on the 
#complete graph 
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',linewidth=20,alpha=0.5)#can use hexcodes as wele here, alpha argumnet tells the amount of transparency of the line.
#linewidth can also be replaced just with *lw*
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',lw=20,alpha=0.5,linestyle='--')#line style can also be replaced with *ls*
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',linewidth=2,alpha=1,marker='+',markersize=20)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=2,alpha=1,marker='o',markersize=20,markerfacecolor='yellow',markeredgewidth=2,markeredgecolor='green')
```
